```python
url = 'https://www.cnbc.com/world/?region=world

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

news_section = soup.find('section', {'class': 'container'})

headlines = []
times = []
news_links = []

for article in news_section.find_all('div', {'class': 'Card-titleArea'}):
    headline = article.find('h3', {'class': 'Card-title'}).text.strip()
    time = article.find('time', {'class': 'Card-time'}).text.strip()
    news_link = article.find('a', {'class': 'Card-titleLink'})['href']
    
    headlines.append(headline)
    times.append(time)
    news_links.append(news_link)

data = {'Headline': headlines, 'Time': times, 'News Link': news_links}
df = pd.DataFrame(data)

print(df)
```

```python

url = 'https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles'

response = requests.get(url)

if response.status_code == 200:
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    articles = soup.find_all('div', class_='pod-listing')
    
    paper_titles = []
    authors_list = []
    published_dates = []
    paper_urls = []
    
    for article in articles:
        
        paper_title = article.find('a', class_='pod-listing-title-link').text.strip()
        paper_titles.append(paper_title)
        
    
        authors = article.find('div', class_='authors').text.strip()
        authors_list.append(authors)
        
    
        published_date = article.find('div', class_='cover-date').text.strip()
        published_dates.append(published_date)
        
    
        paper_url = article.find('a', class_='pod-listing-title-link')['href']
        paper_urls.append(paper_url)
    
    
    df = pd.DataFrame({
        'Paper Title': paper_titles,
        'Authors': authors_list,
        'Published Date': published_dates,
        'Paper URL': paper_urls
    })
    

    print(df)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python

url = 'https://www.dineout.co.in/delhi-restaurants'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

restaurant_names = []
cuisines = []
locations = []
ratings = []
image_urls = []

for restaurant in soup.find_all('div', {'class': 'restnt-info'}):
    name = restaurant.find('a', {'class': 'restnt-info__name'}).text.strip()
    cuisine = restaurant.find('div', {'class': 'restnt-info__cusine'}).text.strip()
    location = restaurant.find('div', {'class': 'restnt-info__rest-location'}).text.strip()
    rating = restaurant.find('div', {'class': 'restnt-rating-popup'}).text.strip()
    image_url = restaurant.find('img')['data-src']
    
    restaurant_names.append(name)
    cuisines.append(cuisine)
    locations.append(location)
    ratings.append(rating)
    image_urls.append(image_url)


data = {
    'Restaurant Name': restaurant_names,
    'Cuisine': cuisines,
    'Location': locations,
    'Ratings': ratings,
    'Image URL': image_urls
}
df = pd.DataFrame(data)

print(df)
```
