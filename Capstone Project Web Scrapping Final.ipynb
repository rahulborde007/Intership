# 1.Austin city Data scrap

```python
# use def function to scrape page
def scrape_page():
    # Extract information
    rental_prices = [price.text.strip() for price in driver.find_elements(By.XPATH, "//p[@class='font-24-b']")]
    addresses = [address.text.strip() for address in driver.find_elements(By.XPATH, "//a[@class='property-card-bottom-address__text link-ghost font-xs one-line-string']")]
    beds = [bed.text.strip() for bed in driver.find_elements(By.XPATH, "//div[@class='labels__item rb-icon-before rb-icon--bedrooms']")]
    bathrooms = [bathroom.text.strip() for bathroom in driver.find_elements(By.XPATH, "//div[@class='labels__item rb-icon-before rb-icon--baths ng-star-inserted']")]
    areas = [area.text.strip() for area in driver.find_elements(By.XPATH, "//div[@class='labels__item rb-icon-before rb-icon--square ng-star-inserted']")]

    # all arrays have the same length
    max_length = max(len(rental_prices), len(addresses), len(beds), len(bathrooms), len(areas))
    rental_prices += [''] * (max_length - len(rental_prices))
    addresses += [''] * (max_length - len(addresses))
    beds += [''] * (max_length - len(beds))
    bathrooms += [''] * (max_length - len(bathrooms))
    areas += [''] * (max_length - len(areas))

    return rental_prices, addresses, beds, bathrooms, areas


def main():
    # Initialize lists to store data
    all_rental_prices = []
    all_addresses = []
    all_beds = []
    all_bathrooms = []
    all_areas = []

    # Scrape data until at least 300 entries are obtained
    while len(all_rental_prices) < 300:
        
        rental_prices, addresses, beds, bathrooms, areas = scrape_page()

        # Add scraped data to the lists
        all_rental_prices.extend(rental_prices)
        all_addresses.extend(addresses)
        all_beds.extend(beds)
        all_bathrooms.extend(bathrooms)
        all_areas.extend(areas)

        # Click the next button
        try:
            next_button = driver.find_element(By.XPATH, "//a[@class='pagination-item']")
            next_button.click()
            time.sleep(20)  
        except Exception as e:
            print("No more pages to load.")
            break


    driver.quit()

    # all arrays have the same length
    max_length = max(len(all_rental_prices), len(all_addresses), len(all_beds), len(all_bathrooms), len(all_areas))
    all_rental_prices += [''] * (max_length - len(all_rental_prices))
    all_addresses += [''] * (max_length - len(all_addresses))
    all_beds += [''] * (max_length - len(all_beds))
    all_bathrooms += [''] * (max_length - len(all_bathrooms))
    all_areas += [''] * (max_length - len(all_areas))

    # Create a dataframe from the scraped data
    data = {
        "Rental Price": all_rental_prices,
        "Address": all_addresses,
        "No of Beds": all_beds,
        "No of Bathrooms": all_bathrooms,
        "Area": all_areas
    }
    Austin_df = pd.DataFrame(data)

    
    print(Austin_df)
    
     # Save the printed data to a CSV file
    output_file = 'austin_data.csv'
    Austin_df.to_csv(output_file, index=False)
    print(f"Data saved to {output_file}")

if __name__ == "__main__":
    main()
```

# Convert all currency in to INR

```python
import re


def convert_to_inr(row):
    # its rate given in project
    conversion_rates = {
        '': 90.73,  
        '$': 83.12,  
        '': 95.73,  
        '': 0.55  
    }
    
    price_str = row['Rental Price'] 
    currency_symbol = price_str[0]  
    
    # Check if the currency symbol is in the conversion_rates dictionary
    if currency_symbol in conversion_rates:
        conversion_rate = conversion_rates[currency_symbol]
        
        numeric_values = [float(val.replace(',', '')) for val in re.findall(r'\d+(?:\.\d+)?', price_str)]
    
        total_price = sum(numeric_values)
        
        return round(total_price * conversion_rate)
    else:
        return None  


data = pd.read_csv("https://raw.githubusercontent.com/rahulborde007/Intership/main/New_dataset.csv")


df = data.assign(**{'Rental Price (INR)': data.apply(convert_to_inr, axis=1)})


print(df)
```

```python
# Function to convert area to sqft
def convert_to_sqft(area):
    if isinstance(area, str):
        if 'Sq Ft' in area:
            return round(float(area.replace(' Sq Ft', '').replace(',', '')), 1)
        elif 'Sq M' in area:
            sq_m = float(area.replace(' Sq M', '').replace(',', ''))
            return round(sq_m * 10.7639, 1)  # 1 square meter = 10.7639 square feet
        else:
            return None
    else:
        return round(area, 1)  # Return the same value if it's not a string

# Apply conversion to the 'Area' column
df['Area'] = df['Area'].apply(convert_to_sqft)

# Print updated DataFrame
print(df)
```
